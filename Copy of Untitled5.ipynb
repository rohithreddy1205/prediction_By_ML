Importing the Dependencies

```python
# Importing all libraries required
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

Collecting and Reading The Data

```python
# Reading data
url = 'https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv'
df = pd.read_csv(url)
print("Data imported successfully")

df.head(10)
```

Data Description

Hours : Continuous Predictor Variable

Score : Continuous Target Variable

Plotting and visualizing the Data

```python
# PLotting the graph of scores
df.plot(x='Hours',y='Scores',style='*',color='blue')
plt.title('Hours vs Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage score')
plt.show()
```


     The above graph clearly shows that there is a positive correlation between hours and percentage

 **Observing The Distribution Of the Continuous Variables**

```python
df.hist(['Hours','Scores'],figsize=(18,5))
```

Checking Missing Values

```python
df.isnull().sum()
```

```python
# We can also use corr to determine the corelation between the variables
df.corr()
```

 **Preparing The Data**

The next step is to divide the data into "attributes" (inputs) and "labels" (outputs).

```python
X=df.iloc[:,:-1].values #all rows and col except last col
Y=df.iloc[:,1].values  #all rows and only 1st col i.e scores
```

```python
X
```

```python
Y
```

**Splitting The Data Into Training and Testing Data**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42000)
```

**Training the Algorithm**

```python
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(X_train, Y_train) 

print("Training complete.")
```

**Plotting The Regression Line**

```python
# Plotting the regression line
line = regressor.coef_*X+regressor.intercept_

# Plotting for the test data
plt.scatter(X, Y)
plt.plot(X, line);
plt.show()
```

**Predicting On Test Data**

Now that we have trained our algorithm, it's time to make some predictions.

```python
print(X_test) # Testing data - In Hours
Y_pred = regressor.predict(X_test) # Predicting the scores
```

**Comparing Actual vs Predicted Results**

```python
# Comparing Actual vs Predicted
ds = pd.DataFrame({'Actual': Y_test, 'Predicted': Y_pred})  
ds
```

**Plotting a bar graph to see the difference between actual and predicted value**

```python
ds.plot(kind='bar',figsize=(5,5))
plt.grid(which='major',linewidth='0.5',color='blue')
plt.grid(which='minor',linewidth='0.5',color='red')
plt.show()
```

**Prediction of score when a student studies for 9.25 hours**

```python
# test with your own data
Hours = [[9.25]] #predict takes 2D array 
own_pred = regressor.predict(Hours)
print("No of Hours = {}".format(Hours))
print("Predicted Score = {}".format(own_pred[0]))
```

**Mean Absolute Error**

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(Y_test, Y_pred))
```

**Conclusion: When a student studies for 9.25 hours per day , the prediction score will be 94%**
